```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")  

model.train(
    data="/kaggle/working/brain_project/data.yaml",
    epochs=120,          
    imgsz=640,
    batch=16,            
    device=0,
    workers=4,           
    cache=True,          
    amp=True,
    optimizer="AdamW",
    lr0=0.001,
    patience=30,
    save=True
)
```
